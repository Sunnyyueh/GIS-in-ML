```python
# Task 1:

import pandas as pd

inputfile= open('cities.txt')

outputfile=open('sample.txt','w')
for each_line in inputfile:
    each_line=each_line.replace("|",",")
    outputfile.write(each_line)

data = pd.read_csv('sample.txt', header = None)
          
dic={}
for i in range(len(data[0])):
    dic[data[0][i]]=(data[2][i],data[3][i])
    

print(dic)
```

```python
#Task 2:
# ! pip install folium
import folium
m = folium.Map(location=[dic["las vegas"][0], dic["las vegas"][1]], tiles = 'cartodbpositron', zoom_start=11, control_scale=True)

for point in dic:
    folium.Marker([dic[point][0],dic[point][1]]).add_to(m)

m
```

```python
from geopy import distance
import itertools as combinations

city=list(dic.keys())
pair=list(combinations(city,2))

dis_output=open('city_distance.txt','w')
for i in range(len(pair)):
    dis_output.write (str(pair[i][0])+"|"+str(pair[i][1])+"|"+str(distance.distance(dic[pair[i][0]],dic[pair[i][1]]).km)+'\n')
    print(str(pair[i][0])+"|"+str(pair[i][1])+"|"+str(distance.distance(dic[pair[i][0]],dic[pair[i][1]]).km))
    
    
```
